```python
import spacy
nlp = spacy.load('en')
doc = nlp('Hello  World!')
for token in doc:
    print(token.text,token.idx)
    
doc = nlp("Next week I'll   be in Madrid.")
list_nnp=[]
token_list=[]
for token in doc:
    print('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t'.format(
        token.text,
        token.idx,
        token.lemma_,
        token.is_punct,
        token.is_space,
        token.shape_,
        token.tag_,
        token.pos_))
    token_list.append(token)
    if token.tag_=='NNP':
        
        list_nnp.append(token.text)
    
print("NNP list")
print(list_nnp)
print("Token list")
print(token_list)
no_nnp=[]
for token in token_list:
    token =str(token)
    
    for nnp in list_nnp:
        if token != nnp:
            no_nnp.append(token)
print("NO NNP LIST")
print(no_nnp)
```
